# GeoChain – Graph Analysis

Rules: the next country must start with the **last letter** of the current country (Polish names).

We model this as a **directed multigraph** where:
- **Nodes** = letters
- **Edges** = countries (directed: `first_letter → last_letter`)

A valid game chain is a **walk** on this graph.

````python
import re
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
from collections import defaultdict

# ── clean names (strip leading space, remove footnote refs) ──────────────────
def clean(name):
    name = name.strip()
    name = re.sub(r'\[\d+\]', '', name)
    # keep only the first "word group" — drop parenthetical additions
    # e.g. "Boliwia" is already clean; "Wyspy Świętego Tomasza i Książęca" stays
    return name.strip()

df['name'] = df['Państwo'].apply(clean)
df['first'] = df['name'].str[0].str.upper()
df['last']  = df['name'].str[-1].str.upper()

df[['name', 'first', 'last']].head(10)
````

## 1 · Letter frequency (first vs last)

````python
all_letters = sorted(set(df['first']) | set(df['last']))

first_counts = df['first'].value_counts().reindex(all_letters, fill_value=0)
last_counts  = df['last'].value_counts().reindex(all_letters, fill_value=0)

x = np.arange(len(all_letters))
width = 0.4

fig, ax = plt.subplots(figsize=(16, 5))
ax.bar(x - width/2, first_counts.values, width, label='starts with (out-edges)', color='steelblue')
ax.bar(x + width/2, last_counts.values,  width, label='ends with   (in-edges)',  color='coral')
ax.set_xticks(x)
ax.set_xticklabels(all_letters, fontsize=11)
ax.set_ylabel('Number of countries')
ax.set_title('Countries starting / ending with each letter')
ax.legend()
plt.tight_layout()
plt.show()
````

## 2 · Dead-end & orphan letters

- **Dead-end letter**: appears as a *last* letter but **no country starts** with it → the game gets stuck after landing on such a country.
- **Orphan letter**: appears as a *first* letter but **no country ends** with it → you can never reach countries starting with it (except at the very beginning).

````python
first_set = set(df['first'])
last_set  = set(df['last'])

dead_end_letters = sorted(last_set - first_set)   # ends here, nothing starts here
orphan_letters   = sorted(first_set - last_set)   # starts here, nothing ends here

print("Dead-end letters (last letter with no country starting there):")
for l in dead_end_letters:
    countries = df[df['last'] == l]['name'].tolist()
    print(f"  {l!r:4s} ({len(countries)} countries): {', '.join(countries)}")

print()
print("Orphan letters (first letter with no country ending there):")
for l in orphan_letters:
    countries = df[df['first'] == l]['name'].tolist()
    print(f"  {l!r:4s} ({len(countries)} countries): {', '.join(countries)}")
````

## 3 · Build the directed multigraph

````python
G = nx.MultiDiGraph()

for _, row in df.iterrows():
    G.add_edge(row['first'], row['last'], country=row['name'])

print(f"Nodes (letters): {G.number_of_nodes()}")
print(f"Edges (countries): {G.number_of_edges()}")
print(f"Letters present: {sorted(G.nodes())}")
````

## 4 · Node degree analysis

Each node's **out-degree** = how many countries start with that letter (choices you have when arriving there).  
Each node's **in-degree** = how many countries end with that letter (how often you land there).

````python
degree_df = pd.DataFrame({
    'letter':     sorted(G.nodes()),
    'out_degree': [G.out_degree(n) for n in sorted(G.nodes())],
    'in_degree':  [G.in_degree(n)  for n in sorted(G.nodes())],
}).assign(balance=lambda d: d['out_degree'] - d['in_degree'])

degree_df = degree_df.sort_values('out_degree', ascending=False).reset_index(drop=True)
print(degree_df.to_string(index=False))
````

````python
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

for ax, col, color, title in [
    (axes[0], 'out_degree', 'steelblue', 'Out-degree: countries you can play next'),
    (axes[1], 'in_degree',  'coral',     'In-degree: countries that land here'),
]:
    sub = degree_df.sort_values(col, ascending=False)
    ax.bar(sub['letter'], sub[col], color=color)
    ax.set_xlabel('Letter')
    ax.set_ylabel('Number of countries')
    ax.set_title(title)
    for bar, val in zip(ax.patches, sub[col]):
        if val > 0:
            ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.3,
                    str(val), ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
````

## 5 · Eulerian path analysis

An **Eulerian path** on the letter-graph would use **every country exactly once**.  
Conditions for a directed Eulerian path:
1. The underlying undirected graph is connected.
2. At most one node has `out − in = +1` (start).
3. At most one node has `in − out = +1` (end).
4. All other nodes have `in == out`.

````python
surplus = {n: G.out_degree(n) - G.in_degree(n) for n in G.nodes()}

starts = [n for n, s in surplus.items() if s == +1]
ends   = [n for n, s in surplus.items() if s == -1]
imbal  = [n for n, s in surplus.items() if abs(s) > 1]

weakly_connected = nx.is_weakly_connected(G)

print(f"Weakly connected: {weakly_connected}")
print(f"Nodes with out-in = +1 (potential starts): {starts}")
print(f"Nodes with in-out = +1 (potential ends):   {ends}")
print(f"Nodes with |imbalance| > 1 (blockers):     {imbal}")
print()

if not imbal and len(starts) <= 1 and len(ends) <= 1 and weakly_connected:
    if starts:
        print("✓ Eulerian PATH exists (use every country exactly once).")
        print(f"  Must start at '{starts[0]}' and end at '{ends[0]}'.")
    else:
        print("✓ Eulerian CIRCUIT exists (use every country, return to start).")
else:
    print("✗ No Eulerian path over all countries.")
    print()
    print("Imbalance per letter (out_degree − in_degree):")
    for n, s in sorted(surplus.items(), key=lambda x: -abs(x[1])):
        bar = '█' * abs(s) if abs(s) > 0 else '·'
        sign = '+' if s > 0 else ''
        print(f"  {n}: {sign}{s:3d}  {bar}")
````

## 6 · Strongly connected components (SCCs)

An SCC is a set of letters you can cycle between indefinitely. Countries whose first *and* last letters both fall inside one large SCC form the "safe core" of the game — players can always continue.

````python
sccs = list(nx.strongly_connected_components(G))
sccs_sorted = sorted(sccs, key=len, reverse=True)

print(f"Total SCCs: {len(sccs_sorted)}\n")
for i, scc in enumerate(sccs_sorted):
    scc_letters = sorted(scc)
    # count countries whose both first and last letter are in this SCC
    in_scc = df[df['first'].isin(scc) & df['last'].isin(scc)]
    print(f"SCC #{i+1}  letters={scc_letters}  ({len(in_scc)} countries fully inside)")
    if len(in_scc) <= 30:
        for name in in_scc['name'].tolist():
            print(f"    {name}")
````

## 7 · Trap countries

A **trap** country ends with a dead-end letter — playing it immediately ends the chain (no country can follow).

````python
traps = df[df['last'].isin(dead_end_letters)].copy()
traps = traps.sort_values('last')

print(f"Trap countries ({len(traps)} total):\n")
for letter, group in traps.groupby('last'):
    print(f"  Ends with '{letter}' (no country starts here):")
    for name in group['name']:
        print(f"    • {name}")
````

## 8 · Longest possible chain (DFS heuristic)

Finding the absolute longest chain is NP-hard (Hamiltonian path), so we use a **greedy DFS with backtracking**, capped at a reasonable depth, to find a very long chain.

````python
from collections import defaultdict
from tqdm import tqdm

# Build adjacency as: letter -> list of (country_name, next_letter)
adj = defaultdict(list)
for _, row in df.iterrows():
    adj[row['first']].append((row['name'], row['last']))

best_chain = []

def dfs(current_letter, used, chain):
    global best_chain
    if len(chain) > len(best_chain):
        best_chain = chain[:]

    # sort options: prefer letters with fewest remaining outgoing options (Warnsdorff heuristic)
    options = [(name, nxt) for name, nxt in adj[current_letter] if name not in used]
    options.sort(key=lambda x: len([c for c, _ in adj[x[1]] if c not in used]))

    for name, nxt in options:
        used.add(name)
        chain.append(name)
        dfs(nxt, used, chain)
        chain.pop()
        used.remove(name)

# try starting from every letter
for start_letter in tqdm(sorted(adj.keys())):
    for start_country, next_letter in adj[start_letter]:
        dfs(next_letter, {start_country}, [start_country])

print(f"Longest chain found: {len(best_chain)} countries\n")
for i, c in enumerate(best_chain, 1):
    row = df[df['name'] == c].iloc[0]
    print(f"  {i:3d}. {c}  ({row['first']} → {row['last']})")
````

## 11 · Letter-to-letter distance matrix

**What it answers:** "If I just named a country ending in letter X, what is the *minimum* number of countries I need to say to reach a state where the last letter is Y?"

Each cell `[X][Y]` = shortest walk length on the letter graph from X to Y.  
- Value `1` means there is a direct country `X → Y`.  
- Value `2` means you need one intermediate letter.  
- `∞` means Y is unreachable from X (dead-end or disconnected).

This matrix is the key planning tool for path generation: before committing to a country, check whether the letter it ends on can still reach the letters you need.

````python
import math

# simple directed graph (letter → letter, weight = 1 per hop)
Gs = nx.DiGraph()
for u, v in G.edges():
    Gs.add_edge(u, v)

letters_ordered = sorted(Gs.nodes())
n_letters = len(letters_ordered)
idx = {l: i for i, l in enumerate(letters_ordered)}

INF = float('inf')
dist_matrix = [[INF] * n_letters for _ in range(n_letters)]

for src in letters_ordered:
    lengths = nx.single_source_shortest_path_length(Gs, src)
    for tgt, d in lengths.items():
        dist_matrix[idx[src]][idx[tgt]] = d

# show as DataFrame
dist_df = pd.DataFrame(dist_matrix, index=letters_ordered, columns=letters_ordered)
dist_df = dist_df.replace(INF, '∞')
print("Letter-to-letter shortest distance (number of countries to say):\n")
print(dist_df.to_string())
````

````python
fig, ax = plt.subplots(figsize=(14, 11))

numeric_mat = [[v if v != INF else np.nan for v in row] for row in dist_matrix]
im = ax.imshow(numeric_mat, cmap='YlOrRd', aspect='auto')

ax.set_xticks(range(n_letters)); ax.set_xticklabels(letters_ordered, fontsize=9)
ax.set_yticks(range(n_letters)); ax.set_yticklabels(letters_ordered, fontsize=9)
ax.set_xlabel('Target letter (last letter you want to reach)')
ax.set_ylabel('Source letter (last letter of previous country)')
ax.set_title('Minimum countries needed to travel from letter → letter\n(grey = unreachable / ∞)')

for i in range(n_letters):
    for j in range(n_letters):
        v = dist_matrix[i][j]
        if v != INF:
            ax.text(j, i, str(int(v)), ha='center', va='center', fontsize=7,
                    color='white' if v >= 4 else 'black')
        else:
            ax.text(j, i, '∞', ha='center', va='center', fontsize=7, color='#aaa')

plt.colorbar(im, ax=ax, label='Hop count (countries)')
plt.tight_layout()
plt.show()
````

## 12 · Path-generation algorithm

### The problem

We want a **valid chain of exactly `target` countries** (e.g. ⌊194 × 2/3⌋ ≈ 130), where:
- No country is repeated.
- Each country's first letter equals the previous country's last letter.

This is a **longest-path variant on a directed multigraph** — NP-hard in general.  
The key insight is that the distance matrix lets us **prune early**: if we are at letter `X` with `k` countries left to place, but the remaining unused countries can only ever form chains shorter than `k`, we can backtrack immediately instead of searching blindly.

### Algorithm: DFS + Warnsdorff + look-ahead pruning

```
def generate_path(adj, dist, target):
    best = []

    def dfs(letter, used, chain):
        # ── early-exit: target reached ──────────────────────────
        if len(chain) == target:
            return chain[:]

        # ── look-ahead pruning ───────────────────────────────────
        # remaining unused countries reachable from `letter`
        reachable = estimate_reachable(letter, used, dist)
        remaining_needed = target - len(chain)
        if reachable < remaining_needed:
            return None   # can't possibly reach target → backtrack

        # ── Warnsdorff move ordering ─────────────────────────────
        # prefer moves whose destination letter has fewest onward options
        # (use tight options first, keep flexible letters for later)
        options = [(c, nxt) for c, nxt in adj[letter] if c not in used]
        options.sort(key=lambda x: onward_count(x[1], used))

        for country, nxt in options:
            result = dfs(nxt, used | {country}, chain + [country])
            if result:
                return result

        return None   # dead end
```

Two heuristics work together:
| Heuristic | What it does | Why it helps |
|---|---|---|
| **Warnsdorff** | Visit letters with fewest remaining exits first | Avoids saving constrained letters for last (where you'd get stuck) |
| **Look-ahead pruning** | Abort branch if reachable count < remaining needed | Cuts the search tree drastically without missing solutions |

`estimate_reachable` is the crucial piece — it asks: *given the letters I'm at and the countries already used, how many more can I chain?* A cheap upper-bound version just counts unused countries whose first letter is reachable from `letter` via the distance matrix.

````python
def generate_path(target, start_letter=None, verbose=False):
    """
    Generate a valid GeoChain path of exactly `target` countries.
    Returns the chain (list of country names), or the longest found if target is unreachable.
    Depends only on `G` (multigraph) and `df` defined earlier in the notebook.
    """
    # ── build letter-level reachability from G ───────────────────────────────
    Gs_local = nx.DiGraph()
    for u, v in G.edges():
        Gs_local.add_edge(u, v)

    _letters = sorted(Gs_local.nodes())
    _idx = {l: i for i, l in enumerate(_letters)}
    _n = len(_letters)
    _INF = float('inf')

    _dm = [[_INF] * _n for _ in range(_n)]
    for src in _letters:
        for tgt, d_val in nx.single_source_shortest_path_length(Gs_local, src).items():
            _dm[_idx[src]][_idx[tgt]] = d_val

    def _dist(src, tgt):
        if src not in _idx or tgt not in _idx:
            return _INF
        return _dm[_idx[src]][_idx[tgt]]

    # ── adjacency: letter → [(country_name, last_letter)] ────────────────────
    _adj = defaultdict(list)
    for _, row in df.iterrows():
        _adj[row['first']].append((row['name'], row['last']))

    # ── upper-bound reachable count from current letter ──────────────────────
    def _reachable(letter, used):
        return sum(
            1 for _, row in df.iterrows()
            if row['name'] not in used and _dist(letter, row['first']) < _INF
        )

    best = [[]]

    def dfs(letter, used, chain):
        if len(chain) > len(best[0]):
            best[0] = chain[:]

        if len(chain) == target:
            return chain[:]

        if _reachable(letter, used) < target - len(chain):
            return None  # prune

        options = [(c, nxt) for c, nxt in _adj[letter] if c not in used]
        # Warnsdorff: prefer moves to letters with fewest remaining exits
        options.sort(key=lambda x: sum(
            1 for c, _ in _adj[x[1]] if c not in used and c != x[0]
        ))

        for country, nxt in options:
            used.add(country)
            result = dfs(nxt, used, chain + [country])
            used.discard(country)
            if result is not None:
                return result

        return None

    start_letters = [start_letter] if start_letter else sorted(_adj.keys())

    result = None
    for sl in start_letters:
        result = dfs(sl, set(), [])
        if result is not None:
            break

    final = result if result is not None else best[0]
    if verbose:
        if len(final) == target:
            print(f"✓ Found chain of exactly {target} countries.")
        else:
            print(f"⚠ Could not reach {target}; longest found: {len(final)} countries.")
    return final
````

````python
def find_bridge(from_letter, to_letter, max_hops=3):
    """
    Find the shortest sequence of countries (length 2 or 3) that travels
    from `from_letter` to `to_letter`.

    Returns a list of country names, or [] if unreachable within max_hops.
    """
    _adj = defaultdict(list)
    for _, row in df.iterrows():
        _adj[row['first']].append((row['name'], row['last']))

    # BFS over (current_letter, path_so_far) — guarantees shortest first
    from collections import deque
    queue = deque()
    queue.append((from_letter, []))

    while queue:
        letter, path = queue.popleft()
        if len(path) > max_hops:
            continue
        for country, nxt in _adj[letter]:
            if country in path:
                continue
            new_path = path + [country]
            if nxt == to_letter:
                return new_path          # found — shortest path
            if len(new_path) < max_hops:
                queue.append((nxt, new_path))

    return []  # unreachable within max_hops


def show_bridge(from_letter, to_letter, max_hops=3):
    bridge = find_bridge(from_letter, to_letter, max_hops)
    if not bridge:
        print(f"  {from_letter} → {to_letter}:  no bridge within {max_hops} hops")
        return
    chain_str = "  →  ".join(
        f"{c}({df[df['name']==c].iloc[0]['last']})" for c in bridge
    )
    print(f"  {from_letter} →[{len(bridge)}]→ {to_letter}:  {chain_str}")


# ── demo: find 2- and 3-hop bridges between selected letter pairs ────────────
print("Bridges of length 2 or 3 between letter pairs:\n")
pairs = [
    ('A', 'N'), ('A', 'E'), ('S', 'B'), ('K', 'F'),
    ('N', 'S'), ('T', 'V'), ('O', 'Z'), ('E', 'K'),
]
for src, tgt in pairs:
    show_bridge(src, tgt, max_hops=3)

print()
print("All reachable pairs with min-hop ≤ 2:")
all_letters = sorted(set(df['first']) | set(df['last']))
for src in all_letters:
    for tgt in all_letters:
        if src == tgt:
            continue
        b = find_bridge(src, tgt, max_hops=2)
        if b:
            print(f"  {src}→{tgt} ({len(b)} hop{'s' if len(b)>1 else ''}): "
                  + " → ".join(b))
````

````python
def validate_chain(chain):
    if not chain:
        print("Empty chain.")
        return
    errors = []
    for i in range(1, len(chain)):
        prev = df[df['name'] == chain[i-1]].iloc[0]
        curr = df[df['name'] == chain[i]].iloc[0]
        if prev['last'] != curr['first']:
            errors.append(f"  Step {i}: '{chain[i-1]}' ends '{prev['last']}'"
                          f" but '{chain[i]}' starts '{curr['first']}'")
    dupes = [c for c in set(chain) if chain.count(c) > 1]
    if dupes:
        errors.append(f"  Duplicates: {dupes}")
    if errors:
        print("INVALID:"); [print(e) for e in errors]
    else:
        fr = df[df['name'] == chain[0]].iloc[0]
        lr = df[df['name'] == chain[-1]].iloc[0]
        print(f"✓ Valid chain of {len(chain)}: "
              f"'{chain[0]}'({fr['first']}…{fr['last']}) → … → "
              f"'{chain[-1]}'({lr['first']}…{lr['last']})")

# validate a few example bridges
for src, tgt in [('A', 'N'), ('S', 'B'), ('T', 'V')]:
    b = find_bridge(src, tgt, max_hops=3)
    validate_chain(b)
````
